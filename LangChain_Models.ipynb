# Open Source Models
Freely avialable AI models that can be downloaded, notified, fine-tuned and deployed without restrictions from central provider.

```python
!pip install -U langchain-huggingface transformers huggingface_hub
```

```python
from langchain_huggingface import ChatHuggingFace, HuggingFaceEndpoint
from langchain_core.messages import HumanMessage
from google.colab import userdata
hf_token = userdata.get('HF_TOKEN')
```

```python
llm = HuggingFaceEndpoint(
    repo_id = "meta-llama/Llama-3.1-8B-Instruct",
    task = "text-generation",
    # provider = "hf-inference",
    huggingfacehub_api_token=hf_token
)
model = ChatHuggingFace(llm=llm)
messages = [
    HumanMessage(content="What is the meaning of life?")
]

result = model.invoke(messages)
print(result)
```

## By Locally running the HuggingFace


```python
langchain.verbose = False
langchain.debug = False
langchain.llm_cache = None

from langchain_huggingface import ChatHuggingFace, HuggingFacePipeline
```

```python
!pip install -U transformers huggingface_hub langchain-huggingface
```

```python
llm = HuggingFacePipeline.from_model_id(
    model_id = "meta-llama/Llama-3.1-8B-Instruct",
    task = 'text-generation',
    pipeline_kwargs = dict(
        temperature = 0.8,
        max_new_tokens = 100)
    )

model = ChatHuggingFace(llm = llm)

result = model.invoke("What is the capital of india?")

print(result)
```
